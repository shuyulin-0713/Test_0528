```python
import numpy as np
import pandas as pd

tips_df = pd.read_csv('tips.csv')
tips_df.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips_df
```

```python
tips_df['小費比例'] = tips_df['小費'] / tips_df['總票價']
tips_df
```

```python
grouped = tips_df.groupby(['日期','吸煙者'])
grouped_pct = grouped['小費']
grouped_pct.mean().unstack('吸煙者')
```

```python
grouped = tips_df.groupby(['日期','吸煙者'])
grouped_pct = grouped[['小費','小費比例']]
grouped_pct.mean().unstack('吸煙者')
```

```python
grouped = tips_df.groupby(['日期','吸煙者'])
grouped_pct = grouped['小費']
grouped_pct.agg(['mean','std'])
```

```python
grouped = tips_df.groupby(['日期','吸煙者'])
grouped_pct = grouped['小費']
grouped_pct.agg([('小費平均','mean'),('小費標準差','std')])
```

```python
grouped = tips_df.groupby(['日期','吸煙者'])
grouped_pct = grouped[['總票價','小費']]
grouped_pct.agg([('平均','mean'),('最大值','max'),('數量','count')])
```

```python
grouped = tips_df.groupby(['日期','吸煙者'])
grouped_pct = grouped[['總票價','小費']]
result = grouped_pct.agg([('平均','mean'),('最大值','max'),('數量','count')])
result['總票價'][['最大值','數量']]
```

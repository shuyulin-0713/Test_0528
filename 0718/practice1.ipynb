```python
import numpy as np
from numpy import ndarray
import pandas as pd
from pandas import DataFrame

scores:ndarray | None = None # 學生的分數
scores_df:DataFrame | None = None # 學生分數的DataFrame
# --------------------------------------------------
# 1. 生成隨機分數資料 (NumPy 陣列)
# --------------------------------------------------
# 隨機生成 50 到 100 之間的整數，形狀為 50 列 5 行 (50位學生、5個科目)
scores = np.random.randint(50, 101, size=(50, 5))

# --------------------------------------------------
# 2. 將 NumPy 陣列轉換成 Pandas DataFrame 資料表
# --------------------------------------------------
scores_df = pd.DataFrame(scores,                             # 傳入剛剛生成的隨機分數數據
            columns=['國文', '英文', '數學', '地理', '歷史'],  # 指定這 5 行的欄位名稱 (科目)
            index=range(1,51))                               # 指定這 50 列的列索引 (座號 1 到 50)
scores_df.T
```

```python
file = open("student.txt","r",encoding="utf-8") #開啟檔案
print(type(file)) #檔案物件的類型
content = file.read() #讀取檔案內容
print(content) #印出檔案內容
file.close() #關閉檔案
file.closed #檢查檔案是否已經關閉
```
